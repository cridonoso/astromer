```python
fs = 20
columns = 1
rows = 4

ids_row = np.random.choice(list(train_kepler.keys()), 1)[0]
ids_row = b'05653152'
print('ID: {}'.format(ids_row))
#np_sample_ids = sample_ids.reshape(rows, columns)

fig, axes = plt.subplots(rows, columns, figsize=(15, 10), squeeze=False, sharex=True)

np_lc = train_kepler[ids_row].numpy()
min_mjd = np.min(train_kepler[ids_row].numpy()[:,0])
max_mjd = np.max(train_kepler[ids_row].numpy()[:,0])

min_flux = np.min(train_kepler[ids_row].numpy()[:,1])
max_flux = np.max(train_kepler[ids_row].numpy()[:,1])

cadence = np.diff(np_lc[:,0])
axes[0, 0].set_title('KEPLER - Num_obs: {} - | Cadence: Median {:.3f}, Mean = {:.3f}, Std = {:.3f}'.format(len(np_lc), np.median(cadence), np.mean(cadence), np.std(cadence)), size=fs)
axes[0, 0].plot(np_lc[:,0], np_lc[:,1], linewidth=1)
axes[0, 0].scatter(np_lc[:,0], np_lc[:,1], s=0.5)
axes[0, 0].tick_params(axis='both', which='major', labelsize=fs)
axes[0, 0].set_ylim(min_flux - 0.05, max_flux + 0.05)
axes[0, 0].set_xlim(min_mjd - 10, max_mjd + 10)


np_lc = train_kepler_alcock_linear[ids_row].numpy()
cadence = np.diff(np_lc[:,0])
#np_lc_filtered = np_lc[np_lc[:,0] < max_mjd]
axes[1, 0].set_title('KEPLER with ALCOCK cadence - Num_obs: {} - | Cadence: Median = {:.3f}, Mean = {:.3f}, Std = {:.3f}'.format(len(np_lc), np.median(cadence), np.mean(cadence), np.std(cadence)), size=fs)
axes[1, 0].plot(np_lc[:,0], np_lc[:,1], linewidth=1)
axes[1, 0].scatter(np_lc[:,0], np_lc[:,1], s=4)
axes[1, 0].tick_params(axis='both', which='major', labelsize=fs)
axes[1, 0].set_ylim(min_flux - 0.05, max_flux + 0.05)
axes[1, 0].set_xlim(min_mjd - 10, max_mjd + 10)

np_lc = train_kepler_ogle_linear[ids_row].numpy()
cadence = np.diff(np_lc[:,0])
#np_lc_filtered = np_lc[np_lc[:,0] < max_mjd]
axes[2, 0].set_title('KEPLER with OGLE cadence - Num_obs: {} - | Cadence: Median = {:.3f}, Mean = {:.3f}, Std = {:.3f}'.format(len(np_lc), np.median(cadence), np.mean(cadence), np.std(cadence)), size=fs)
axes[2, 0].plot(np_lc[:,0], np_lc[:,1], linewidth=1)
axes[2, 0].scatter(np_lc[:,0], np_lc[:,1], s=4)
axes[2, 0].tick_params(axis='both', which='major', labelsize=fs)
axes[2, 0].set_ylim(min_flux - 0.05, max_flux + 0.05)
axes[2, 0].set_xlim(min_mjd - 10, max_mjd + 10)

np_lc = train_kepler_atlas_linear[ids_row].numpy()
cadence = np.diff(np_lc[:,0])
#np_lc_filtered = np_lc[np_lc[:,0] < max_mjd]
axes[3, 0].set_title('KEPLER with ATLAS cadence - Num_obs: {} - | Cadence: Median = {:.3f}, Mean = {:.3f}, Std = {:.3f}'.format(len(np_lc), np.median(cadence), np.mean(cadence), np.std(cadence)), size=fs)
axes[3, 0].plot(np_lc[:,0], np_lc[:,1], linewidth=1)
axes[3, 0].scatter(np_lc[:,0], np_lc[:,1], s=4)
axes[3, 0].tick_params(axis='both', which='major', labelsize=fs)
axes[3, 0].set_ylim(min_flux - 0.05, max_flux + 0.05)
axes[3, 0].set_xlim(min_mjd - 10, max_mjd + 10)
    
plt.tight_layout()
plt.show()
```
