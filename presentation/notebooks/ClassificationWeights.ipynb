```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np

import glob
import os

from tensorflow.keras.losses import CategoricalCrossentropy
from presentation.pipelines.steps.model_design import load_pt_model, build_classifier 
from presentation.pipelines.steps.load_data import build_loader 

pd.options.mode.chained_assignment = None  # default='warn'
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

%load_ext autoreload
%autoreload 2
```

```python
root = './presentation/results/diagstromer/2024-12-02_14-13-12/'
```

```python
n_classes = {'alcock': 6, 'atlas': 5}
```

```python
results = {'alcock': {}, 'atlas':{}}
for ds_name in ['alcock', 'atlas']:
    for spc in [20, 100, 500]:
        fold_gammas = []
        for fold_n in range(3):
        
            ft_case = '{}/fold_{}/{}_{}'.format(ds_name, fold_n, ds_name, spc)
            curr = os.path.join(root, 'classification', ft_case, 'skip_avg_mlp')

            astromer, model_config = load_pt_model(os.path.join(root, 'finetuning', ft_case))

            model = build_classifier(astromer, 
                                    model_config, 
                                    False, 
                                    n_classes[ds_name],
                                    arch='skip_avg_mlp')

            model.load_weights(os.path.join(curr, 'weights'))
            gamma_values = model.get_layer('gamma_weight').gamma.numpy()
            fold_gammas.append(gamma_values)

        results[ds_name][spc] =  fold_gammas
        
    
```

```python
fig, axes = plt.subplots(3, 2, figsize=(5, 5), dpi=300, sharex='col', sharey='row', 
                         gridspec_kw={'hspace':0.05, 'wspace':0.05})
x = [r'$\gamma_0$', r'$\gamma_1$', r'$\gamma_2$', r'$\gamma_3$', r'$\gamma_4$', r'$\gamma_5$', r'$\gamma_6$']

names  = ['Alcock', 'ATLAS']
for i, dsname in enumerate(['alcock', 'atlas']):
    for k, spc in enumerate([20, 100, 500]):
        mean_values = np.mean(results[dsname][spc], 0)
        std_values = np.std(results[dsname][spc], 0)
        
        axes[k][i].bar(range(len(x)),mean_values, yerr=std_values, 
                    color='#E1D5E7', ecolor='k', edgecolor='#9673A6')
        
        axes[k][i].set_xticks(range(len(x)))
        axes[k][i].set_xticklabels(x)

        for j, v in enumerate(mean_values):
            axes[k][i].text(j-0.45, v+0.004, '{:.2f}'.format(v), fontsize=6, color='black')

        ax_twin = axes[k][-1].twinx()
        ax_twin.text(7.4, 0.2, '{} SPC'.format(spc), ha='right', fontsize=8, rotation=270)  
        ax_twin.set_yticks([])
        if spc == 20:
            supv = 0.165
            minv = 0.12
        if spc == 100:
            supv = 0.175
            minv = 0.12
        if spc == 500:
            supv = 0.185
            minv = 0.1
            
        axes[k][i].set_ylim(minv, supv)
        axes[k][i].set_yticks([])
        # axes[k][i].set_yticks([minv, supv])
        # axes[k][i].set_yticklabels([minv, supv], fontsize=8)   
        
    axes[0][i].set_title(names[i], fontsize=8)
    
axes[1][0].set_ylabel('Weight Value', fontsize=8)


plt.savefig('./presentation/figures/classification/gamma_weights.pdf', format='pdf', 
            bbox_inches='tight')

plt.show()
```
