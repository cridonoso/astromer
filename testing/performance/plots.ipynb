```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('time_memory.csv',index_col=[0])
x = list(np.arange(df.shape[0])+1)
data_1 = df['Time[s]']
data_2 = df['Memory[MB]'] 
fig, ax1 = plt.subplots() 
  
ax1.set_xlabel('# Cores') 
ax1.set_ylabel('Time [s]') 
ax1.plot(x, data_1, 'o-r', color = 'red', label ='Time (s)'  ) 
plt.legend() 
# Adding Twin Axes
ax1.legend(loc=9)
ax2 = ax1.twinx() 
  
ax2.set_ylabel('Memory [MB]') 
ax2.plot(x, data_2, 'o-r',color = 'blue' , label ='Memory [MB]') 
ax2.legend(loc=0)
# Show plot
plt.legend()
plt.show()
```
